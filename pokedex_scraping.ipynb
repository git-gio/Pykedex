```python
# import the necessary packages
from bs4 import BeautifulSoup
import requests
import os
```

```python
dirname = os.path.dirname(__file__)
utils_dir = os.path.join(dirname, 'utils/pokemon_list_gen1-9.txt')

# construct the soup and initialize the list of pokemon names
soup = BeautifulSoup(open(utils_dir).read())
names = []

# loop over all link elements
for link in soup.findAll("a", {"class": "ent-name"}):
    # update the list of pokemon names
    names.append(link.text)
```

```python
print(names)
for name in names:
    # initialize the parsed name as just the lowercase
    # version of the pokemon name
    parsedName = name.lower()
    # if the name contains an apostrophe (such as in
    # Farfetch'd, just simply remove it)
    parsedName = parsedName.replace("'", "")
    # if the name contains a period followed by a space
    # (as is the case with Mr. Mime), then replace it
    # with a dash
    parsedName = parsedName.replace(". ", "-")
    # handle the case for Nidoran (female)
    if name.find(u'\u2640') != -1:
        parsedName = name + "-f"
        print(name + " --> " + parsedName)
    # and handle the case for Nidoran (male)
    elif name.find(u'\u2642') != -1:
        parsedName = name + "-m"
        print(name + " --> " + parsedName)

    # construct the URL to download the sprite
    print("downloading %s" % (name))
    url = "http://img.pokemondb.net/sprites/yellow/normal/%s.png" % (parsedName)
    #url = "https://play.pokemonshowdown.com/sprites/xyani/%s.gif" % (parsedName) # alternati download for all pokemons
    r = requests.get(url)
    # if the status code is not 200, ignore the sprite
    if r.status_code != 200:
        print("error downloading %s" % (name))
        continue
    # write the sprite to file
    f = open("%s/%s.png" % ("sprites/sprites_gen3", name.lower()), "wb")
    f.write(r.content)
    f.close()
```
